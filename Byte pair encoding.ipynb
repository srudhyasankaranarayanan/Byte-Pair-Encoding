```python
import numpy as np
```

```python
# Read the corpus
with open("corpus.txt", "r") as file:
    words = [line.strip().lower() for line in file if line.strip()]

print("Corpus:")
print(words)
```

```python
# Create a vocabulary
vocab = {}

for word in words:
    text = " ".join(word) + " </w>"

    if text in vocab:
        vocab[text] += 1
    else:
        vocab[text] = 1

print("Initial Vocabulary")

for word, freq in vocab.items():
    print(word, ":", freq)
```

```python
# Frequency Statistics
freq = np.array(list(vocab.values()))

print("Word Frequencies:", freq)
print("Total Words:", np.sum(freq))
print("Maximum Frequency:", np.max(freq))
```

```python
# Functions to count the pair
def get_pairs(vocab):

    pairs = {}

    for word, count in vocab.items():

        letters = word.split()

        for i in range(len(letters) - 1):

            pair = (letters[i], letters[i + 1])

            if pair in pairs:
                pairs[pair] += count
            else:
                pairs[pair] = count

    return pairs
```

```python
# Functions to merge the pair
def merge(pair, vocab):

    new_vocab = {}

    old = " ".join(pair)
    new = pair[0] + pair[1]

    for word, count in vocab.items():

        word = word.replace(old, new)

        new_vocab[word] = count

    return new_vocab
```

```python
# Perform BPE Merges
rules = []

for i in range(10):

    pairs = get_pairs(vocab)

    if len(pairs) == 0:
        break

    best = max(pairs, key=pairs.get)

    print("Merge", i + 1)
    print("Best Pair:", best)
    print("Frequency:", pairs[best])

    rules.append(best)

    vocab = merge(best, vocab)
```

```python
# Final Vocabulary
final_vocab = set()

for word in vocab:

    for token in word.split():

        final_vocab.add(token)

print("Final Vocabulary:")
print(sorted(final_vocab))

print("Vocabulary Size:", len(final_vocab))
```

```python
# Encode Function
def encode(word):

    tokens = list(word) + ["</w>"]

    for pair in rules:

        i = 0

        while i < len(tokens) - 1:

            if tokens[i] == pair[0] and tokens[i + 1] == pair[1]:

                tokens[i:i+2] = [pair[0] + pair[1]]

            else:
                i += 1

    return tokens
```

```python
# Decode Function
def decode(tokens):

    word = "".join(tokens)

    return word.replace("</w>", "")
```

```python
# Test the model
test_words = ["newest", "lowest", "lower"]

for word in test_words:

    encoded = encode(word)

    decoded = decode(encoded)

    print("\nOriginal:", word)
    print("Encoded :", encoded)
    print("Decoded :", decoded)
```

```python
# Display the merge cell
print("Learned Merge Rules")

for i, rule in enumerate(rules, 1):

    print(i, ":", rule)

print("\nProgram Completed")
```
